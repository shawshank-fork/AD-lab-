```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = pd.DataFrame({
    "Area": [800, 1000, 1200, 1500, 1800, 2000, 2200, 2500],
    "Rooms": [2, 3, 3, 4, 4, 5, 5, 6],
    "Price": [150, 200, 240, 300, 360, 400, 440, 500]  
})

X = data[["Area", "Rooms"]]
y = data["Price"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("R2 Score:", r2)

print("\nCoefficients:", model.coef_)
print("Intercept:", model.intercept_)

plt.scatter(data["Area"], data["Price"], label="Actual Data")
area_range = np.linspace(data["Area"].min(), data["Area"].max(), 100)
rooms_mean = data["Rooms"].mean()
pred_line = model.predict(
    pd.DataFrame({"Area": area_range, "Rooms": [rooms_mean]*100})
)

plt.plot(area_range, pred_line, label="Regression Line")

plt.xlabel("Area")
plt.ylabel("Price")
plt.title("House Price Prediction (Linear Regression)")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
x = np.array([8,12,11,5,14,3,6,8,4,9]).reshape(-1,1)
y = np.array([25,35,29,24,38,12,18,27,17,30])

model = LinearRegression()
model.fit(x, y)

slope = model.coef_[0]
intercept = model.intercept_

print("Using sklearn:")
print("Slope (m):", slope)
print("Intercept (c):", intercept)

y_pred = model.predict(x)
print("Predicted values:", y_pred)
x_flat = x.flatten() 

x_mean = np.mean(x_flat)
y_mean = np.mean(y)

numerator = np.sum((x_flat - x_mean) * (y - y_mean))
denominator = np.sum((x_flat - x_mean) ** 2)

m = numerator / denominator
c = y_mean - m * x_mean

print("\nManual Calculation:")
print("Slope (m):", m)
print("Intercept (c):", c)
plt.scatter(x, y, label="Actual Data")
plt.plot(x, y_pred, label="Regression Line")

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear Regression")
plt.legend()
plt.grid()
plt.show()
```
